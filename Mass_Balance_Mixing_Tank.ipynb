### Mixing Tank Mass Balance Original version

```python
print("\n === Mixing Tank Mass Balance ===")
print("\n")

# Inputs from user
F1 = float(input("Flow Rate of Stream 1 (F1):)"))
C1 = float(input("Concentration of Stream 1 (C1):)"))
F2 = float(input("Flow Rate of Stream 2 (F2):)"))
C2 = float(input("Concentration of Stream 2 (C2):)"))

# Total Flow Rate
F_out = F1 + F2

# Total Concentration
C_out = (F1*C1 + F2*C2)/F_out

# Output
print (f"\n Total Outlet Flow Rate: {F_out:.2f} L/min ")
print (f"\n Outlet Concentration: {C_out:.4f} mol/L " )
```

### Mixing Tank Mass Balance (List Input)

```python
F1 = flow_rate_of_stream_1 = 25
C1 = concentration_of_stream_1 = 10
F2 = flow_rate_of_stream_2 = 30
C2 = concentration_of_stream_2 = 15

# Inputs from a list
data = {'Flow Rate (L/min)': [F1, F2],
        'Concentration (mol/L)': [C1, C2]}
# data = {'Flow Rate (L/min)': [25.0, 10.0],
#         'Concentration (mol/L)': [30.0, 15.0]}
input_df = pd.DataFrame(data, index=['Stream 1', 'Stream 2'])

display(input_df)

# Total Flow Rate
F_out = input_df['Flow Rate (L/min)'].sum()

# Total Mass Flow Rate of component
Total_Mass_Flow = (input_df['Flow Rate (L/min)'] * input_df['Concentration (mol/L)']).sum()

# Total Concentration
C_out = Total_Mass_Flow / F_out

# Output
print (f"\n Total Outlet Flow Rate: {F_out:.2f} L/min ")
print (f"\n Outlet Concentration: {C_out:.4f} mol/L " )
```

### Mixing Tank Mass Balance (List of Lists Input for n Streams)

```python
# Inputs as a list of lists for n streams: [[F1, C1], [F2, C2], ..., [Fn, Cn]]
stream_data = [
              [25.0, 10.0],
              [30.0, 15.0],
              [00.0, 0.0],
              [14.0, 15.25]
]

# Create a list of stream names for the index
stream_names = [f'Stream {i+1}' for i in range(len(stream_data))]

# Create the input_df DataFrame
input_df = pd.DataFrame(stream_data, columns=['Flow Rate (L/min)', 'Concentration (mol/L)'], index=stream_names)

display(input_df)

# Total Flow Rate
F_out = input_df['Flow Rate (L/min)'].sum()

# Total Mass Flow Rate of component
Total_Mass_Flow = (input_df['Flow Rate (L/min)'] * input_df['Concentration (mol/L)']).sum()

# Total Concentration
C_out = Total_Mass_Flow / F_out

# Output
print (f"\n Total Outlet Flow Rate: {F_out:.2f} L/min ")
print (f"\n Outlet Concentration: {C_out:.4f} mol/L " )
```
